```python
with open('./inputs/1.txt', 'r') as file:
  dial_pos = 50
  zero_pointed = 0
  second_zero_pointed = 0
  for line in file:
    raw_move = line.strip()
    dir, clicks = (1 if raw_move[0] == 'R' else -1, int(raw_move[1:]))
    dist_to_zero = 100 - dial_pos if dir == 1 else \
      dial_pos if dial_pos > 0 else 100

    if clicks >= dist_to_zero:
      second_zero_pointed += 1 + (clicks - dist_to_zero) // 100

    tmp_calc = dial_pos + clicks * dir
    dial_pos = ((tmp_calc % 100) + 100) % 100

    if dial_pos == 0:
      zero_pointed += 1

  print('Part 1:', zero_pointed)
  print('Part 2:', second_zero_pointed)
```
